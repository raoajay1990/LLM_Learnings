```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")
```

```python
word_vectors = model
word_vectors['computer]
```

```python
word_vectors.similarity('woman','man')
word_vectors.similarity('boy', 'girl')
word_vectors.similarity('neice', 'wall')
```

```python
word_vectors.most_similar('tower', topn = 3)
```
